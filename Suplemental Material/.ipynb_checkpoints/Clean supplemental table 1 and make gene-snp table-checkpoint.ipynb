```python
import pandas as pd
from gprofiler import GProfiler
import re
```

```python
gp = GProfiler(return_dataframe=True)
```

```python
df = pd.read_csv("Supplemental Table 1 Identified OCD Genes.csv")
df
```

```python
gwas_df = pd.read_csv('gwas-association-downloaded_2021-03-25-EFO_0004242.tsv', sep = '\t')
gwas_df = gwas_df[['MAPPED_GENE', 'SNP_ID_CURRENT']]
gwas_df
```

```python
new_rows = []
for index, row in df.iterrows():
    keeps = ['Source', 'Hormonal?', 'Annotation', 'GENE', 'SNP']
    cleaned_row = dict()
    snps = set()
    for label in row.index:                
        if label == 'GENE':
            cleaned_row[label] = row[label]
            for gene in gwas_df.MAPPED_GENE:
                gene = str(gene)
                row_gene = str(row.GENE)
                if row_gene in gene or gene in row_gene or row_gene == gene:
                    gene_gwas_df = gwas_df[gwas_df.MAPPED_GENE == gene]
                    for snp in gene_gwas_df.SNP_ID_CURRENT:
                        snp = str(snp)
                        snp = 'rs' + snp
                        snps.add(snp)
        elif label == 'SNP':
            snp = str(row[label])
            for s in snp.split(','):
                s = s.strip()
                snps.add(s)
        else:
            snp = re.findall(r'rs[0-9]+', str(row[label]))
            if snp:
                for s in snp:
                    snps.add(s)
                if label in keeps:
                    row[label] = float('NaN')
            else:
                if label in keeps:
                    cleaned_row[label] = row[label]
    cleaned_row['SNP'] = snps
    new_rows.append(cleaned_row)
df = pd.DataFrame(new_rows)
df.to_csv('Supplemental_Table_1_gene_list.csv')    
```

```python
rows = []
for index, row in df.iterrows():
    snps = row['SNP']
    if 'nan' in snps:
        snps.remove('nan')
    if len(snps) > 0: 
        for snp in snps:
            new_row = {}
            new_row['GENE'] = row.GENE
            new_row['SNP'] = snp
            rows.append(new_row)
    else:
        new_row = {}
        new_row['GENE'] = row.GENE
        new_row['SNP'] = snp
        rows.append(new_row)
gene_snp_df = pd.DataFrame(rows)
gene_snp_df.to_csv('Gene_SNP_ID_table.csv')
gene_snp_df
```
