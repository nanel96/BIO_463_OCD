```python
import pandas as pd
from gprofiler import GProfiler
```

```python
df = pd.read_csv("Supplemental Table 1 Identified OCD Genes.csv")
df
```

```python
len(pd.unique(df.GENE))
```

```python
gene_list = list(df.GENE)
```

```python
gp = GProfiler(return_dataframe=True)
results_df = gp.profile(organism = 'hsapiens', query = gene_list, no_evidences = False)
results_df = results_df[results_df.source == 'KEGG']
print(len(results_df))
```

```python
results_df
```

```python
results_df.to_csv('kegg_pathway_enrichment.csv')
```

```python
interesting_pathways = ['Serotonergic synapse', 'Glutamatergic synapse', 'Dopaminergic synapse']
results_df = results_df[results_df.name.isin(interesting_pathways)]
results_df = results_df[['native', 'name', 'p_value', 'intersections']]
results_df.columns = ['Pathway_ID', 'Pathway_name', 'p_value', 'enriched_genes']
results_df.to_csv('intersting_kegg_pathways.csv')
results_df
```

```python
gene_set = set()
for gene_list in results_df.enriched_genes:
    for gene in gene_list:
        gene_set.add(gene)
print(gene_set)
```

```python
snp_df = pd.read_csv('Gene_SNP_ID_table.csv')
hormone_pathway_snps = snp_df[snp_df.GENE.isin(gene_set)]
hormone_pathway_snps.to_csv('Hormone_pathway_SNPs.csv')
```
